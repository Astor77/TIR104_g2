```python
import pandas as pd
import json


file_path = "/workspaces/TIR104_g2/D_raw_data/tmdb_detail_raw.json"
with open(file_path, "r", encoding="utf-8") as f:
    details = json.load(f)
```

```python
# 逐一取出對應劇情分類
# df_movie_genres 含 tmdb_id, 劇情分類id, 劇情分類名稱表
movie_genres_list = []
for movie in details:
    id = movie["id"]
    for dict in movie["genres"]:
        dict["tmdb_id"] = id
        movie_genres_list.append(dict)

df_movie_genres = pd.DataFrame(movie_genres_list)
df_movie_genres = df_movie_genres.rename(columns={"id": "genres_id", "name": "genres_name"})
df_movie_genres = df_movie_genres.loc[:, ["tmdb_id", "genres_id", "genres_name"]]
```

```python
# df_details 總表
df_details = pd.DataFrame(details)
df_details.loc[:, ["id", "imdb_id", "title", "release_date", "budget", "revenue"]]
```

```python
# 以下是merge後的呈現
# 可以看要不要merge，或是就各自單獨一張表

details_merge_genres = df_details.merge(
    df_movie_genres,
    how = "left",
    left_on = "id",
    right_on = "tmdb_id",
)

details_merge_genres.loc[:, ["id", "imdb_id", "title", "release_date", "genres_id", "genres_name", "budget"]]
```
