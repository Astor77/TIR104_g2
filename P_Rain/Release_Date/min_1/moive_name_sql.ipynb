```python
import pymysql
import pandas as pd

host = 'host.docker.internal'
port = 3306
user = 'root'
passwd = 'password'
db = 'movie'
charset = 'utf8mb4'

conn = pymysql.connect(host=host, port=port, user=user, passwd=passwd, db=db, charset=charset)
print('Successfully connected!')

cursor = conn.cursor()

# DELETE ALL

sql = """
DELETE FROM `movie_name`;
"""
cursor.execute(sql)
conn.commit()

print('Successfully delete!')

# INSERT 

# 讀取檔案
file_path = "v1_mapping_close_true.csv" #正確檔案路徑
df = pd.read_csv(file_path)

# 生成 SQL INSERT 語句
table_name = "movie_name"
sql_statements = []

for _, row in df.iterrows():
    sql = f"""
    INSERT INTO `{table_name}` (`Year`, `MovieId`, `Name`, `tmdb_id`, `tmdb_title`)
    VALUES ({row['Year']}, {row['MovieId']}, '{row['Name'].replace("'", "''")}', {row['tmdb_id']}, '{row['tmdb_title'].replace("'", "''")}');
    """.strip()
    print(sql)
    cursor.execute(sql)

# 將指令放進 cursor 物件，並執行
conn.commit()

print('Successfully insert!')

cursor.close()
conn.close()
```
